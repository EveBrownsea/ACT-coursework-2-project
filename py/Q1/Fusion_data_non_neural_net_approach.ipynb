This method will be focusing on using a GBR to predict whether or not a fusion experiment will result in ignition. I will be testing whether this version will actually work or if it's a bad approach to the problem I want to solve.


```python
#List of all libraries needed to run this code.
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import GradientBoostingRegressor
import os
import kagglehub
```

```python
#Accessing and loading the data as 'fusion'.
path = kagglehub.dataset_download("adebusayoadewunmi/nuclearfusion-data") + '/fusion_experiment.csv'
print("Path to dataset files:", path)
fusion = pd.read_csv(path)
fusion = pd.get_dummies(fusion, columns = ['Magnetic Field Configuration', 'Target Composition'], prefix='', prefix_sep='')
fusion.drop(['Unnamed: 0'], axis=1)
```

```python
#Creating the train-test split for the data.

#Start by dropping the Ignition column so that it can be set as the target of
#the gbr.
X, y = fusion.drop('Ignition', axis=1), fusion['Ignition']
#Starting off with size=0.2 as that's what's been shown to work in notebooks.
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, shuffle=False)

#Creating and fitting the model:
gbr = GradientBoostingRegressor()
gbr.fit(X_train, y_train)
```
